```python
from ultralytics import YOLO
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
```

```python
# Download the model YOLOv8
model = YOLO('yolov8n.pt')
```

```python


def detect_objects(image_path):
    image = cv2.imread(image_path)
    if image is None:
        raise FileNotFoundError(f"Image not found at path: {image_path}")
    
    # Перетворюємо зображення в формат, який підтримується моделлю YOLOv8
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = model.predict(image_rgb)
    
    return results

def remove_background(image_path, results):
    image = cv2.imread(image_path)
    if image is None:
        raise FileNotFoundError(f"Image not found at path: {image_path}")
    
    mask = np.zeros(image.shape[:2], dtype=np.uint8)
    
    for result in results[0].boxes.xyxy:
        x1, y1, x2, y2 = map(int, result[:4])
        mask[y1:y2, x1:x2] = 255

    foreground = cv2.bitwise_and(image, image, mask=mask)
    background = cv2.bitwise_not(mask)
    result_image = cv2.add(foreground, cv2.cvtColor(background, cv2.COLOR_GRAY2BGR))

    return result_image

def recognize_clothing_items(results):
    clothing_items = []
    for result in results[0].boxes.xyxy:
        x1, y1, x2, y2 = map(int, result[:4])
        cls = int(result[5]) if len(result) > 5 else -1
        conf = result[4] if len(result) > 4 else 0
        item = {'class': cls, 'confidence': conf, 'bounding_box': (x1, y1, x2, y2)}
        clothing_items.append(item)
    return clothing_items

```
